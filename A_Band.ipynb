```
import math
import rasterio
import matplotlib.pyplot as plt
from google.colab import drive
#drive.mount('/content/drive/My Drive/merged_image.tif')

image_file = "/content/drive/My Drive/merged_image.tif"

sat_data = rasterio.open(image_file)

width_in_projected_units = sat_data.bounds.right - sat_data.bounds.left
height_in_projected_units = sat_data.bounds.top - sat_data.bounds.bottom


print("Width: {}, Height: {}".format(width_in_projected_units, height_in_projected_units))
print("Rows: {}, Columns: {}".format(sat_data.height, sat_data.width))

#Upper left pixel
row_min = 0
col_min = 0

#Lower right pixel.  
row_max = sat_data.height - 1
col_max = sat_data.width - 1

topleft = sat_data.transform * (row_min, col_min)
bottomright = sat_data.transform * (row_max, col_max)

print("Top left corner coordinates: {}".format(topleft))
print("Bottom right corner coordinates: {}".format(bottomright))

print(sat_data.count)
print(sat_data.indexes)


# Load the 5 bands into 2d arrays 
b, g, r, n, ni = sat_data.read()

# Display the blue band.

fig = plt.imshow(b)
plt.show()

# Display the green band.

fig = plt.imshow(g)
fig.set_cmap('gist_earth')
plt.show()

# Displaythe red band.

fig = plt.imshow(r)
fig.set_cmap('inferno')
plt.colorbar()
plt.show()

# Display the infrared band.

fig = plt.imshow(n)
fig.set_cmap('winter')
plt.colorbar()
plt.show()
```
